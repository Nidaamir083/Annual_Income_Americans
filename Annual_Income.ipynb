```python
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_data = one_hot_encoder.fit_transform(df[['occupation']])

encoded_df = pd.DataFrame(encoded_data, columns=one_hot_encoder.get_feature_names_out(['occupation']))

df_encoded = pd.concat([df, encoded_df], axis=1)
print(df_encoded)
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_data = one_hot_encoder.fit_transform(df[['marital.status']])

encoded_df = pd.DataFrame(encoded_data, columns=one_hot_encoder.get_feature_names_out(['marital.status']))

df_encoded = pd.concat([df, encoded_df], axis=1)
print(df_encoded)
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_data = one_hot_encoder.fit_transform(df[['education']])

encoded_df = pd.DataFrame(encoded_data, columns=one_hot_encoder.get_feature_names_out(['education']))

df_encoded = pd.concat([df, encoded_df], axis=1)
print(df_encoded)
```

```python
df_encoded
```

```python
df_encoded.drop(columns=['education','marital.status','occupation'],inplace=True)
```

```python
df_encoded
```

```python
import tensorflow as tf

# 1. Prepare the category-to-index mapping
unique_categories = df_encoded['native.country'].unique()
category_to_index = {category: idx for idx, category in enumerate(unique_categories)}

# 2. Encode the 'native.country' column
df_encoded['native.country_encoded'] = df_encoded['native.country'].map(category_to_index)

# 3. Create the embedding layer
embedding_dim = 5  # Choose embedding size
embedding_layer = tf.keras.layers.Embedding(input_dim=len(unique_categories), output_dim=embedding_dim)

# 4. Get embeddings for the encoded column
inputs = tf.constant(df_encoded['native.country_encoded'].values)
embeddings = embedding_layer(inputs)

# 5. Add embeddings to the DataFrame (optional)
# You can add the embeddings as new columns if you want:
embedding_df = pd.DataFrame(embeddings.numpy())
embedding_df.columns = [f'native.country_embedding_{i}' for i in range(embedding_dim)]
df_encoded = pd.concat([df_encoded, embedding_df], axis=1)

# 6. Drop the original 'native.country' column (optional)
# If you no longer need the original column:
df_encoded.drop(columns=['native.country'], inplace=True)

print(df_encoded)
```

```python
df_encoded
```

```python
X = df_encoded.drop(columns=['income_encoded'])
y = df_encoded['income_encoded'].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(34,activation='sigmoid',input_dim=34))
model.add(Dense(24,activation='sigmoid'))
model.add(Dense(12,activation='sigmoid'))
model.add(Dense(8,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train_scaled,y_train,epochs=30,verbose=1,validation_split=0.2)
```

```python
model.layers[2].get_weights()
```

```python
ylog = model.predict(X_test_scaled)
```

```python
ylog[0]
```

```python
y_pred = np.where(ylog>0.5,1,0)
```

```python
y_pred[9]
```

```python
y_pred[4]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
